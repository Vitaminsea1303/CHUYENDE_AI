#Yêu cầu 2

```python

# Các cột còn lại để so sánh
columns_to_compare = df.drop(columns=['target']).columns

# 1. Tính toán thống kê cơ bản
print("Thống kê cơ bản theo nhóm target:")
stats = df.groupby('target')[columns_to_compare].describe().transpose()
print(stats)
```

```python
sns.set_style('whitegrid')
sns.histplot(df['target'],color ='red', bins = 10)
```

```python


# 2. Vẽ biểu đồ boxplot cho từng cột
plt.figure(figsize=(16, 20))
for i, col in enumerate(columns_to_compare, 1):
    plt.subplot(len(columns_to_compare) // 2 + 1, 2, i)
    sns.boxplot(data=df, x='target', y=col, palette='Set2')
    plt.title(f'Comparison of {col} between Groups', fontsize=14)
    plt.xlabel('Heart Disease (0 = No, 1 = Yes)', fontsize=12)
    plt.ylabel(col, fontsize=12)

plt.tight_layout()
plt.show()

# 3. Vẽ histogram để so sánh phân phối của từng cột
plt.figure(figsize=(16, 20))
for i, col in enumerate(columns_to_compare, 1):
    plt.subplot(len(columns_to_compare) // 2 + 1, 2, i)
    sns.histplot(df[df['target'] == 0][col], kde=True, color='blue', label='No Heart Disease (target=0)', bins=15)
    sns.histplot(df[df['target'] == 1][col], kde=True, color='red', label='Heart Disease (target=1)', bins=15)
    plt.title(f'Distribution of {col} by Heart Disease', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.legend()

plt.tight_layout()
plt.show()
```

#Mối liên hệ


```python

from scipy.stats import pearsonr
# 1. Tính hệ số tương quan
correlation, p_value = pearsonr(df['age'], df['max_heart_rate'])
print(f"Hệ số tương quan Pearson giữa tuổi và nhịp tim tối đa: {correlation:.2f}")
print(f"p-value: {p_value:.4f}")

# 2. Vẽ biểu đồ phân tán với đường hồi quy
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='max_heart_rate', hue='target', palette='Set1', alpha=0.7)
sns.regplot(data=df, x='age', y='max_heart_rate', scatter=False, color='blue', label='Trend Line')
plt.title('Relationship between Age and Max Heart Rate', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Max Heart Rate', fontsize=14)
plt.legend(title='Heart Disease (Target)')
plt.grid(alpha=0.3)
plt.show()
```

Đánh giá
Đường xu hướng (trend line) nghiêng xuống, cho thấy khi tuổi tác tăng lên thì nhịp tim tối đa có xu hướng giảm. Điều này phù hợp với kiến thức sinh lý học về sự lão hóa của hệ tim mạch.
Các điểm dữ liệu phân bố tương đối đều xung quanh đường xu hướng, tuy nhiên vẫn có một số điểm lệch xa. Điều này cho thấy tuổi tác không phải là yếu tố duy nhất ảnh hưởng đến nhịp tim tối đa, mà còn có thể chịu tác động của các yếu tố khác như giới tính, tình trạng sức khỏe, hoạt động thể chất..
Phân loại theo nguy cơ mắc bệnh tim: Biểu đồ sử dụng màu sắc để phân biệt các cá nhân có nguy cơ mắc bệnh tim (Target = 1) và không có nguy cơ (Target = 0). Tuy nhiên, từ biểu đồ này, chúng ta chưa thể kết luận rõ ràng về mối liên hệ giữa nguy cơ mắc bệnh tim và nhịp tim tối đa. Có thể cần phải phân tích sâu hơn bằng các phương pháp thống kê khác.
=>Biểu đồ cho thấy một xu hướng giảm dần của nhịp tim tối đa theo tuổi tác. Tuy nhiên, để đưa ra kết luận chính xác và đầy đủ hơn, cần tiến hành phân tích thống kê sâu hơn, xem xét thêm các yếu tố ảnh hưởng và xây dựng các mô hình dự đoán.
